### 3.1.4 평균값(기대값)

### 3.1.5 표본분산

* 데이터가 평균값에서 얼마나 떨어져 있는지 나타내는 지표

$$\sigma^{2} = {1 \over N} \sum_{i=1}^{N}\left ( x_{i} - \mu \right )^{2}$$

````python
sigma_2_sample = np.sum((fish_data - mu) ** 2)/N
sigma_2_sample
````

- 위의 공식을 아래와 같이 단계별로 나누어 결과를 확인하면서 식 전체에 대한 이해 도모

````python
fish_data
````

````python
fish_data - mu
````

````python
(fish_data -mu)**2
````

````python
np.sum((fish_data - mu)**2)
````

````python
sp.var(fish_data, ddof = 0)
````

````python
np.var(fish_data, ddof = 0)
````

- 여기서 ddof = 0 인 인자는 Delta Degrees of Freedom으로 표본분산을 구할 때 자유도의 값을 제하고 나누어주는 항인 1/(N - df)의 계산에서 df 값을 지정하는 인수이다.  

### 3.1.6 불편분산

* 우리가 모집단의 분산을 추정하기 위한 추정량으로 표본분산을 가져온다면, 표본분산의 평균은 실제 분산보다 (n-1)/n 배 만큼 차이가 발생

    - 추정량이 편향되지 않게 조정해주고자 표본분산에 n/(n-1)만큼을 곱하여 분산의 비편향추정량을 얻으므로 표본분산의 불편추정량은 n이 아니라 n-1로 나누어줌
    
$$\sigma^{2} = {1 \over N-1} \sum_{i=1}^{N}\left ( x_{i} - \mu \right )^{2}$$
> (참고) https://towardsdatascience.com/why-sample-variance-is-divided-by-n-1-89821b83ef6d

> (참고) https://blog.naver.com/krov98/222697830711

````python
sigma_2 = np.sum((fish_data - mu) ** 2) / (N-1)
````

````python
sigma_2
````

````python
np.var(fish_data, ddof = 1)
````

### 3.1.7 표준편차

$$\sigma = \sqrt{\sigma^{2}} = \sqrt{{1 \over N-1} \sum_{i=1}^{N}\left ( x_{i} - \mu \right )^{2} }$$

````python
sigma = np.sqrt(sigma_2)
````

````python
sigma
````

````python
np.std(fish_data, ddof = 1)
````

### 3.2.7 공분산(cov)

* **공분산** : 2개 연속형 변수의 관계성을 확인하는 통계량

    - **공분산이 0보다 클때** : 변수 한 쪽이 커지면 다른 한 쪽도 커짐
    - **공분산이 0보다 작을 때** : 변수 한 쪽이 큰 값을 갖게 되면 다른 한 쪽은 작아짐
    - **공분산이 0일 때** : 변수 사이의 관계성 없음
    
$$ Cov(x, y) = \frac{1}{N} \sum_{N}^{i = 1} \left ( x_i - \mu_x \right )\left ( y_i - \mu_y \right ) $$

    
$$ Cov(x, y) = \frac{1}{N-1} \sum_{N}^{i = 1} \left ( x_i - \mu_x \right )\left ( y_i - \mu_y \right ) $$

### 3.2.8 분산-공분산 행렬

* 여러 변수의 분산과 공분산을 행렬 형태로 정리한 것 

$$Cov(x, y) = \begin{bmatrix}
\sigma_{x}^{2} & Cov(x, y) \\
Cov(x, y) & \sigma_{y}^{2} \\
\end{bmatrix}$$

### 3.2.9 공분산(실습)

````python
cov_data = pd.read_csv("py_source/3-2-3-cov.csv")
print(cov_data)
````

* 공분산을 계산하는 데 사용할 요소들을 계산하여 값에 저장

````python
# 데이터 분리
x = cov_data["x"]
y = cov_data["y"]

# 표본크기
N = len(cov_data)

# 평균값 계산
mu_x = np.mean(x)
mu_y = np.mean(y)
````

* 공분산 계산

````python
cov_sample = sum((x - mu_x) * (y -mu_y)) / N
cov_sample
````

분모를 N-1로 설정

````python
cov = sum((x - mu_x) * (y -mu_y)) / (N - 1)
cov
````

### 3.2.10 분산-공분산 행렬(실습)

````python
sp.cov(x, y, ddof = 0)
````

````python
np.cov(x, y, ddof = 0)
````

* 자유도 1 지정 후 계산

### 3.2.11 피어슨 상관계수

* 공분산의 표준화 값 (-1 ~ 1 사이로 표준화)

$$\rho_xy = \frac{Cov(x, y)}{\sqrt{\sigma_{x}^{2}\sigma_{y}^{2}}}$$

### 3.2.12 상관행렬

* 상관행렬은 여러 변수의 상관계수들을 행렬 형태로 정리한 것

$$Corr(x, y) = \begin{bmatrix}
1 & \rho_xy \\
\rho_xy & 1 \\
\end{bmatrix}$$

### 3.2.13 피어슨 상관계수(실습)

````python
# 분산 계산
sigma_2_x = np.var(x, ddof = 1)
sigma_2_y = np.var(y, ddof = 1)

# 상관계수
rho = cov/np.sqrt(sigma_2_x * sigma_2_y)
rho
````

````python
sp.corrcoef(x, y)
````

````python
np.corrcoef(x, y)
````

````python
# 수치 계산에 사용하는 라이브러리
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

# 그래프를 그리기 위한 라이브러리
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

# 표시 자릿수 지정
%precision 3

# 그래프를 주피터 노트북에 그리기 위한 설정
%matplotlib inline
````

### 3.5.7 표본평균을 몇 번이고 계산하는 함수 만들기

* 표본평균을 계산하는 함수 만들기

````python
def calc_sample_mean(size, n_trial):
    sample_mean_array = np.zeros(n_trial)
    for i in range(0, n_trial):
        sample = population.rvs(size = size)
        sample_mean_array[i] = np.mean(sample)
    return(sample_mean_array)
````

````python
np.random.seed(1)
np.mean(calc_sample_mean(size = 10, n_trial = 10000))
````

### 3.5.8 샘플사이즈를 바꿨을 때 표본평균의 분산

* 샘플사이즈를 10, 20, 30으로 변화시켰을 때 표본평균의 분포를 바이올린 플롯을 이용해서 확인

````python
np.random.seed(1)

size_10 = calc_sample_mean(size = 10, n_trial = 10000)
size_10_df = pd.DataFrame({
    "sample_mean" : size_10,
    "size"        : np.tile("size 10", 10000)})


size_20 = calc_sample_mean(size = 20, n_trial = 10000)
size_20_df = pd.DataFrame({
    "sample_mean" : size_20,
    "size"        : np.tile("size 20", 10000)})


size_30 = calc_sample_mean(size = 30, n_trial = 10000)
size_30_df = pd.DataFrame({
    "sample_mean" : size_30,
    "size"        : np.tile("size 30", 10000)})


sim_result = pd.concat([size_10_df, size_20_df, size_30_df])

print(sim_result)
````

* 샘플사이즈별로 데이터 프레임 생성. 각각의 데이터프레임은 시행횟수가 10000번이므로 만개의 행이 데이터 프레임에 들어 있음

* 이 데이터를 이용해서 바이올린 플랏을 그림

### 3.5.9 표본평균의 표준편차는 모집단보다 작다

* **대상**  : 표본평균의 표준편차

* **변화시키는것** : 샘플사이즈

* **알고 싶은 것** : 샘플사이즈가 커질수록 표본 평균의 표준편차는 작아진다 -> 샘플사이즈가 커지면 표본평균은 신뢰할 수 있는 값이 된다.

* 2 ~ 100까지 2씩 차이 나게 하여 샘플사이즈를 준비함

**샘플사이즈를 크게 하면 흩어짐이 적은 신뢰할 수 있는 표본평균을 얻을 수 있음**


### 3.5.10 표준오차

* **표준오차(Standard Error)** : 표본평균의 표준편차 

$$ {SE} = \frac{\sigma}{\sqrt{N}} $$

````python
# 표준오차 공식 적용

standard_error = 0.8/np.sqrt(size_array)
````

### 3.5.15 불편성

* 추정량의 기댓값이 진짜 모수(모집단의 파라미터)가 되는 특성을 불편성이라고 함

* 불편성이 있다는 것은 평균을 냈을 때 과대 또는 과소하지 않는다는 뜻이며, 이는 곧 편향이 없는 추정량이라고 할 수 있다.

### 3.5.16 일치성

* 샘플사이즈가 커지면 추정량이 진짜 모수에 가까워지는 특성을 일치성이라고 함

* 일치성이 있다는 것은 샘플사이즈가 무한할 때 추정량과 모수가 일치한다는 뜻

### 3.5.17 모수에 대해 좋은 추정량

* 정규분포의 모수는 평균과 분산 2개. 정규분포의 모수는 보통 알려지지 않은 상태이므로 표본평균과 불편분산을 대신 사용하게 됨

* 표본평균의 평균값은 모평균과 거의 같은 불편성을 가지고 있으며 샘플 사이즈를 크게 하면 표본평균은 모평균에 가까워지는 일치성을 가지고 있는 것을 확인함

* 불편분산의 평균값 또한 모분산과 거의 같고, 샘플 사이즈를 크게 하면 불편분산이 모분산에 가까워지는 것을 확인함

* 이처럼 불편성과 일치성 모두 가지고 있는 추정량이 모수에 대해 좋은 추정량이라고 할 수 있다.


### 3.5.18 큰수의 법칙

* 큰수의 법칙은 표본의 크기가 커지면 표본평균이 모평균에 가까워지는 방법을 표현한 법칙

### 3.5.19 중심극한정리

* 모집단의 분포가 무엇이든 간에 샘플사이즈가 커지면 확률변수의 합은 정규분포에 가까워짐 -> 중심극한정리

* 동전던지기 시뮬레이션을 통해 확인

````python
n_size = 10000
n_trial = 50000

coin = np.array([0, 1])

count_coin = np.zeros(n_trial)

np.random.seed(1)
for i in range(0, n_trial):
    count_coin[i] = np.sum(
        np.random.choice(coin, size = n_size, replace = True)
    )
````

* **distplot** , **displot** , **histplot** 으로 비교

````python
sns.distplot(count_coin, color = "steelblue")
````

* 표본 평균을 계산할 때는 합계를 구하는 계산이 들어감. 때문에 표본평균은 정규분포에 가까운 분포가 된다고 가정할 수 있음

* 이런 성질은 매우 편리하기 때문에 정규분포는 통계학의 여러 곳에서 모습을 나타낸다. 하지만 어디까지나 표본의 합계치가 정규분포를 따른다는 것이므로 주의를 기울여서 사용해야 한다. 

    * 모집단분포가 이항분포라면 샘플사이즈가 무한히 크다고 해도 모집단이 이항분포인 건 변하지 않기 때문임
    
    * 중심극한정리는 평균값 등의 추정값을 처음 구할 때 사용할 수 있음
    
    * 모집단분포가 정규분포가 아닌 경우에는 일반선형모델 등을 사용할 필요가 있다.
